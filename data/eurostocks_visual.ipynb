## Plot monthly data
Retrieved from https://sdw.ecb.europa.eu/quickview.do?SERIES_KEY=143.FM.M.U2.EUR.DS.EI.DJES50I.HSTA

```python
monthly_df = pd.read_csv("eurostocks/sdw_monthly.csv")
```

```python
monthly_df["y"] = monthly_df["s1"]
monthly_df.index = pd.to_datetime(monthly_df["date"])
```

```python
monthly_df
```

```python
plot_series(monthly_df["y"])
```

```python
# to compare with daily (which starts at 2007)
plot_series(monthly_df["y"][monthly_df["date"] > "2007-01-01"])
```
